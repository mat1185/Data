<h3>Basic python 3 notebook plotting zip codes for Los Angeles county using census data which can be found <a href='https://www.census.gov/geo/reference/zctas.html'>here</a></h3>

```python
#begin by importing basemap to plot our map
import matplotlib.pyplot as plt
import matplotlib.cm
 
from mpl_toolkits.basemap import Basemap
from matplotlib.patches import Polygon
from matplotlib.collections import PatchCollection
from matplotlib.colors import Normalize
```

```python
#create our figure to draw the map
fig, ax = plt.subplots(figsize=(10,20))
#create the map
m = Basemap(resolution = 'h', #high resolution
            projection = 'merc', # mercator projection
            lat_0=33.882670, lon_0=-118.328467, #Coordinates of Los Angeles
            llcrnrlon=-118.6682, llcrnrlat=33.7037, urcrnrlon=-118.150, urcrnrlat=34.3373)#coordinates of the boundaries of the map
m.drawmapboundary(fill_color='white')
#m.drawlsmask(land_color='coral',ocean_color='aqua',lakes=True, resolution='h')
m.drawcoastlines()
m.drawcounties()
m.readshapefile('/data/zt06_d00', 'zip')
```
